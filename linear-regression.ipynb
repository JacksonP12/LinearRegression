```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv(r'C:\Users\gemki\Storage\data-science\Final Project\carbon-emissions.csv',encoding='cp1252')
df
```

```python
years = []
for col in df.columns:
    if col != "State":
        years.append(col)
year = []
for i in years:
    y = int(i)
    year.append(y)

cey = []
for i in year:
    sm = df[str(i)].sum()
    cey.append(int(sm))
```

```python
d = {'years' : year, 'emissions' : cey}
ce_data = pd.DataFrame(data=d)

plt.figure(figsize = (10,5))
plt.ylim(4000000000,6500000000)
plt.ylabel('Carbon Emissions (Metric Tons)')
plt.xlabel('Year')
plt.ticklabel_format(style='plain')
plt.xticks([1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2019])
plt.title('United States Carbon Emissions by Year (Metric Tons)')
plt.scatter(year,cey)
```

```python
import statsmodels.api as sm
import numpy as np
x_ary = year
x_with_cnst = sm.add_constant(x_ary)
model = sm.OLS(cey, x_with_cnst)
results = model.fit()
b,m = results.params
b = float(b)
m = float(m)
print(m)
print(b)
```
